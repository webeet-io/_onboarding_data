# NYC High School Directory:

```python
## Importing the CSV files by suing pandas library.

import pandas as pd

pd.set_option('display.max_columns',None)

df = pd.read_csv('/Users/msubhanzk/Downloads/2. Data Analysis Course/14. June 2025/1. Internship Documents/2. Day 2/1. Dataset/high-school-directory.csv')
df
```

```python
## Want to see the column names:

df.columns
```

```python
## Replacing the columns'name (Clean the columns'name)

new_col_names=[]
for col in df.columns:
    stripped_names= col.lower().strip().replace(' ','_')
    new_col_names.append(stripped_names)
df.columns = new_col_names
df.head()
```

```python
## Want to fillter all the schools located in Brooklyn:
pd.set_option('display.max_columns', None)
brooklyn_schools = df[df['borough']=='Brooklyn']
brooklyn_schools.head() 
```

```python
## Answer to qusetions "How many schools are in Brooklyn?""

num_schools_Brooklyn=df[df['borough']=='Brooklyn'].shape[0]
print('Number of schools in Brooklyn:', num_schools_Brooklyn)
```

```python
## Answer to questions "How many offer Grade 9 entery?"

school_grade_9 = df[df['grade_span_min']== 9]
num_grade9_schools = school_grade_9.shape[0]
print("Number of schools offering Grade 9 entry:", num_grade9_schools)
```

```python
## Want to find out the number of school in each location

num_school_each_location = df.groupby('borough').size().reset_index(name = 'number_of_schools')
num_school_each_location.head()
```

```python
## Create a barchart to show the number of school in each location:

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6)) 
sns.barplot(x = "borough", y = "number_of_schools", data=num_school_each_location)

plt.ylabel("number_of_schools")
plt.xlabel("borough")
plt.title ("Number of schools in each location")
plt.show()
```

```python
## Answer to question "Average number of students per borough?""

avg_num_stud_borough = df.groupby('borough')['total_students'].mean().round(2).reset_index(name = 'Avg_num_students')
avg_students_columns=['borough','Avg_num_of_Student']
avg_num_stud_borough.head()
```

```python
## Answer to question " Summary fo grade_span_max grouped by borough"


sum_grade_span_max = df.groupby('borough')['grade_span_max'].sum().reset_index(name = 'summary_grade_span_max')
sum_grade_span_max.head()
```

```python
## Want to know how many schools do we have in zip codes: 20529.0

total_schools = []
total_schools=df[df['zip_codes']==20529.0].shape[0]
print("Number of schools in zip code 20529.0:",total_schools)
```
