```python
#1: School Distribution

#SQL query in triple quotes to make it a Python string
query_borough_count = """
SELECT
    borough,
    COUNT(*) AS number_of_schools
FROM
    nyc_schools.high_school_directory
GROUP BY
    borough;
"""

#Execute the SQL query using pandas and display the result
df_borough_count = pd.read_sql(query_borough_count, conn)
print(df_borough_count)
```

```python
#2 Language Learners.
#Average % of English Language Learners in Borough

query = """
Select hd.borough,Avg(sd.ell_percent) AS avg_ell_precentage
From nyc_schools.high_school_directory As hd
LEFT JOIN nyc_schools.school_demographics As sd
On hd.dbn = sd.dbn
Group By hd.borough;"""

df = pd.read_sql(query, conn)
df.head()
```

```python
#3 School supporting special needs
# Top 3 Schools with hightest percentage of special education

query = """SELECT borough, school_name, sped_percent
FROM (
    SELECT
        hd.borough,
        hd.school_name,
        sd.sped_percent,
        ROW_NUMBER() OVER (PARTITION BY hd.borough ORDER BY sd.sped_percent DESC) as rank
    From nyc_schools.high_school_directory As hd
    LEFT JOIN nyc_schools.school_demographics As sd
    ON hd.dbn = sd.dbn
    WHERE sd.sped_percent IS NOT NULL
) As ranked
WHERE rank <= 3
ORDER BY borough, rank""";

df = pd.read_sql(query, conn)
df.head()
```
