```python
# Step 5 : Average ELL% per borough
query = """
SELECT d.borough, 
       ROUND(AVG(s.ell_percent)::numeric, 2) AS avg_ell_percent
FROM nyc_schools.school_demographics AS s
JOIN nyc_schools.high_school_directory AS d
ON s.dbn = d.dbn
GROUP BY d.borough
ORDER BY avg_ell_percent DESC;
"""

df_ell = pd.read_sql(query, conn)
df_ell['avg_ell_percent'] = df_ell['avg_ell_percent'].round(2)
df_ell

```

```python
# Step 7: Top 3 Special Needs Schools per Borough

query_demo = "SELECT dbn, sped_percent FROM nyc_schools.school_demographics;"
df_demo = pd.read_sql(query_demo, conn)

query_dir = "SELECT dbn, school_name, borough FROM nyc_schools.high_school_directory;"
df_dir = pd.read_sql(query_dir, conn)

df_merged = pd.merge(df_dir, df_demo, on="dbn", how="inner")

df_merged['sped_percent'] = pd.to_numeric(df_merged['sped_percent'], errors='coerce')


df_top3 = (df_merged
           .sort_values(['borough', 'sped_percent'], ascending=[True, False])
           .groupby('borough')
           .head(3)
           .reset_index(drop=True))


df_top3['sped_percent'] = df_top3['sped_percent'].round(2)

df_top3
```
