## 8. Create Visualization
- Bar chart: Number of schools per borough
- Brooklyn has the highest number of schools compared to other boroughs, but the average number of students per school is the lowest.

```python
## Visualization 
! pip install matplotlib seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## bar chart: Number of schools per borough
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="borough", order=df["borough"].value_counts().index)
plt.title("Number of Schools per Borough")
plt.xlabel("Borough")
plt.ylabel("Number of Schools")
plt.show()
```

```python
## Bar chart: Average number of students per borough
avg_students = df.groupby("borough")["total_students"].mean().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=avg_students, x="borough", y="total_students", order=avg_students.sort_values("total_students", ascending=False)["borough"])
plt.title("Average Number of Students per Borough")
plt.xlabel("Borough")
plt.ylabel("Average Number of Students")
plt.show()
```

## 9. Some Interesting Findings
- CTE schools are the most common type in Brooklyn, while Specialized Schools are the least common.
- Brooklyn offers the most language classes with 26 unique languages, whereas Staten Island offers the fewest language classes (10) with only 8 unique languages.

```python
## Find the school types in. Staten Island and Brooklyn 
Brooklyn_school_types = df[df["borough"]=="Brooklyn"]["school_type"].value_counts()
Brooklyn_school_types.head()

```

```python
##Visualization of school types in Brooklyn
plt.figure(figsize=(10,6))
sns.barplot(x=Brooklyn_school_types.index, y=Brooklyn_school_types.values)
plt.title("School Types in Brooklyn")
plt.xlabel("School Type")
plt.ylabel("Number of Schools")
plt.xticks(rotation=45)
plt.show()
```

```python
## Find the school types in Staten Island
Statden_school_types=df[df["borough"]=="Staten Island"]["school_type"].value_counts()
Statden_school_types.head()
```

```python
## Visualization of school types in Staten Island
plt.figure(figsize=(10,6))
sns.barplot(x=Statden_school_types.index, y=Statden_school_types.values)
plt.title("School Types in Staten Island")
plt.xlabel("School Type")
plt.ylabel("Number of Schools")
plt.xticks(rotation=45)
plt.show()
```
