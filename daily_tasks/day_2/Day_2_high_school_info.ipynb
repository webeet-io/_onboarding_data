```python
#Q4: Filter for Brooklyn
brooklyn_schools = df[df['borough'] == 'Brooklyn']

#Count unique school names
unique_brooklyn_schools = brooklyn_schools['school_name'].nunique()

print(f"Total unique schools in Brooklyn: {unique_brooklyn_schools}")
```

```python
#Q4: Filter for Brooklyn schools
brooklyn_schools = df[df['borough'] == 'Brooklyn']

# Find schools where grade_span_min = 9 (they include grade 9)
grade9_schools = brooklyn_schools[brooklyn_schools['grade_span_min'] == 9]

# Count unique schools offering grade 9
grade9_count = grade9_schools['school_name'].nunique()

print(f"Number of Brooklyn schools offering Grade 9: {grade9_count}")
```

```python
#Q5: Group By borough and count schools
borough_counts = df['borough'].value_counts()
borough_counts
```

```python
#Q5:
avg_students = df.groupby('borough')['total_students'].mean()

#Rounding
avg_students = avg_students.round(0).astype(int)

#Show result
avg_students
```

```python
#Q5: summary of grade_span_max in borough

df.groupby('borough')['grade_span_max'].describe()
```

```python
#Q6: Bar chart of Number of schools per borough

import matplotlib.pyplot as plt

# Assuming 'borough_counts' was calculated in the preceding code:
# borough_counts = df['borough'].value_counts()

borough_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Schools per Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Schools')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Viz 1: bar chart of total students in borough

import matplotlib.pyplot as plt
# Group by borough and sum the total students
total_students_per_borough = df.groupby('borough')['total_students'].sum()

# Create a bar chart
total_students_per_borough.plot(kind='bar', figsize=(10, 6))
plt.title('Total Students per Borough')
plt.xlabel('Borough')
plt.ylabel('Total Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Viz 2: Chart for top 5 schools in a selected borough by total students

import matplotlib.pyplot as plt

# Select a borough, for example, 'Brooklyn'
selected_borough = 'Brooklyn'

# Filter the DataFrame for the selected borough
borough_df = df[df['borough'] == selected_borough].copy()

# Sort by 'total_students' in descending order and get the top 5
top_5_schools = borough_df.sort_values(by='total_students', ascending=False).head(5)

# Create a bar chart
plt.figure(figsize=(12, 7))
plt.bar(top_5_schools['school_name'], top_5_schools['total_students'])
plt.title(f'Top 5 Schools by Total Students in {selected_borough}')
plt.xlabel('School Name')
plt.ylabel('Total Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Viz 3:
df['school_type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of School Types')
plt.ylabel('')
plt.show()
```
