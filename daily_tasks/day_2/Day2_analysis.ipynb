```python
import pandas as pd
```

```python
df = pd.read_csv("high-school-directory.csv") #reading the dataset
```

```python
df.head()
```

```python
df.info()
```

Clean the column names (make lowercase, replace spaces with _, remove special characters)

```python
df.columns = df.columns.str.lower().str.replace(" ", "_") #cleaned column names
```

```python
print(df.columns) #updated column names
```

Filter the dataset to include only schools located in Brooklyn

```python
brooklyn_df = df[df['borough'] == 'Brooklyn']
brooklyn_df.head()
```

How many total schools are in Brooklyn?

```python
total_schools_brooklyn = brooklyn_df['dbn'].nunique() #remove duplicate
print("Total number of schools in Brooklyn:", total_schools_brooklyn)
```

How many offer Grade 9 entry?

```python
grade9_entry_brooklyn = brooklyn_df[brooklyn_df['grade_span_min'].astype(str).str.contains('9.0', na=False)].shape[0]
print("Number of schools in Brooklyn offering Grade 9 entry:", grade9_entry_brooklyn)
```

Count of schools per borough

```python
school_per_borough = df['borough'].value_counts() #count of schools
print(school_per_borough)
```

Average number of students per borough

```python
avg_students_per_borough = round(df.groupby('borough')['total_students'].mean(),2) #average number of students
print(avg_students_per_borough)
```

Summary of grade_span_max grouped by borough

```python
grade_span_max_summary = df.groupby('borough')['grade_span_max'].describe() #summary of grade_span_max
print(grade_span_max_summary)
```

Bar chart: Number of schools per borough

```python
import matplotlib.pyplot as plt
```

```python
colors = ['blue', 'coral', 'green', 'gold', 'purple']

plt.figure(figsize=(10, 6))
ax = school_per_borough.plot(kind='bar', color=colors)
plt.title('Number of Schools per Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Schools')
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(school_per_borough):
    ax.text(i, v + 1, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
colors = ['blue', 'coral', 'green', 'gold', 'purple']

plt.figure(figsize=(10, 6))
ax = avg_students_per_borough.plot(kind='bar', color=colors)
plt.title('Average number of students per borough')
plt.xlabel('Borough')
plt.ylabel('Avg no. of Students')
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(avg_students_per_borough):
    ax.text(i, v + 1, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Observations:
The number of schools per borough are highest in Brooklyn and then follows Bronx. However, the average number of students are highest in Staten Island and then in Queens.
