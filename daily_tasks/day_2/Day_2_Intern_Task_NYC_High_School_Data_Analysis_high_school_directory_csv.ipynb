```python
# Clean column names: lowercase, replace spaces with _, remove special characters
df.columns = (
    df.columns
    .str.lower()
    .str.replace(' ', '_')
    .str.replace(r'[^\w_]', '', regex=True)
)
```

```python
df.columns.tolist()
```

```python
# Check if data has any spacing issues
print("Borough values with quotes to see spaces:")
for borough in df['borough'].unique():
    print(f"'{borough}'")
```

```python
# This handles spaces AND partial matches
brooklyn_schools = df[df['borough'].str.upper().str.contains('BROOKLYN', na=False)]
```

```python
# Check results
print(f"Total schools in dataset: {len(df)}")
print(f"Schools in Brooklyn: {len(brooklyn_schools)}")
```

```python
# Answer to Question
print(f"\nAnswer: There are {len(brooklyn_schools)} total schools in Brooklyn")
```

```python
# First, let's see what the grade columns look like
print("Sample grade_span_min values:")
print(brooklyn_schools['grade_span_min'].unique())
print("\nSample grade_span_max values:")
print(brooklyn_schools['grade_span_max'].unique())
```

```python
grade_9_entry = brooklyn_schools[brooklyn_schools['grade_span_min'] <= 9]
print(f"Brooklyn schools offering Grade 9 entry: {len(grade_9_entry)}")
```

```python
# Count schools per borough (using the full dataset, not just Brooklyn)
schools_per_borough = df['borough'].value_counts()
print("Count of schools per borough:")
print(schools_per_borough)
```

```python
# Average students per borough
avg_students_per_borough = df.groupby('borough')['total_students'].mean()
print("\nAverage number of students per borough:")
print(avg_students_per_borough)
```

```python
# Summary of grade_span_max by borough
grade_max_summary = df.groupby('borough')['grade_span_max'].describe()
print("\nSummary of grade_span_max grouped by borough:")
print(grade_max_summary)
```

```python
# Create bar chart for number of schools per borough
plt.figure(figsize=(10, 6))
sns.barplot(x=schools_per_borough.index, y=schools_per_borough.values, palette='viridis')
plt.title('Number of Schools per Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Schools')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Key Insights from the Analysis

print("KEY INSIGHTS:")
print("=" * 50)

print("1. Brooklyn dominates NYC high school landscape:")
print("   - Brooklyn has the most schools (121), closely followed by Bronx (118)")
print("   - This suggests Brooklyn serves a large student population")

print("\n2. School size varies dramatically by borough:")
print("   - Staten Island schools are largest (avg 1,847 students)")
print("   - Bronx schools are smallest (avg 490 students)")
print("   - This could indicate different educational models or population density")

print("\n3. Grade 12 is the standard across all boroughs:")
print("   - Almost all schools end at Grade 12 (traditional high school model)")
print("   - Very few alternative school structures exist")
```
