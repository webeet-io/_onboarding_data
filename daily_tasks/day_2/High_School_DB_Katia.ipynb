```python
g_i_s = scd[(scd['bus'].notna()) & (scd['subway'].notna())] ### good infrastructure schools where both no null

g_i_s
```

```python
g_i_s[['school_name', 'bus', 'subway']].head(25) ###having the name for each one where school served by bus and subway
```

```python
g_i_s['bus_count'] = g_i_s['bus'].fillna('').apply(lambda x: len(x.split())) ###creates new columns for bus and subway counts
g_i_s['subway_count'] = g_i_s['subway'].fillna('').apply(lambda x: len(x.split()))

filtered = g_i_s[(g_i_s['bus_count'] > 3) & (g_i_s['subway_count'] > 3)][['borough', 'school_name', 'bus', 'subway']] ###bus and subway greater than 3


filtered
```

#357 schools have both services, and 334 schools are served by more than 3 bus lines and by more than 3 subway lines

```python
schoolXborough = filtered.groupby('borough')['school_name'].count() ###popular borough with more than 3 bus lines and 3 subway lines

schoolXborough_sorted = schoolXborough.sort_values(ascending=False)

schoolXborough_sorted
```

```python
schoolXborough_sorted.plot(kind='bar') ###Number of schools per borough

plt.title('Schools x Borough')
plt.tight_layout()
plt.show()
```

```python
grouped_student_per_borough = scd.groupby('borough')['total_students'].mean().sort_values(ascending=False)

grouped_student_per_borough  ###Average number of students per borough
```

While one might expect that a borough with more schools would naturally have more students, this is not always true.

In Brooklyn, the relationship appears to be directly proportional â€” a high number of schools corresponds to a high number of students.

However, this trend does not hold in all boroughs.
For example:

The Bronx has fewer schools than Queens, but more total students.

This suggests that school size (students per school) varies significantly across boroughs.

```python
def weighted_mean(x):
    return (x['grade_span_max'] * x['total_students']).sum() / x['total_students'].sum()

# the weighted-grade-mean (considering a fixed amount of students each borough)
weighted_means = scd.groupby('borough').apply(weighted_mean)

weighted_means
```
