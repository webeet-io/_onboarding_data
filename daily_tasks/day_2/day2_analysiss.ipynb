```python

pd.set_option('display.max_rows', None) # to increase console size

missing = df.isnull().sum()  #to check the null values in the dataset and their %
missing_percentage = (missing / len(df)) * 100
print(pd.concat([missing, missing_percentage.round(2)], axis=1, keys=["Missing", "%"]).sort_values(by="%", ascending=False))


```

## Cleaning the dataset

```python
# standardize the columns
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
```

```python
df.info()
```

## Handling missing data

```python
# Drop columns with >80% missing
high_missing = missing_percentage[missing_percentage > 80].index.tolist()
df.drop(columns=high_missing, inplace=True)
```

```python
df_dropna = df.dropna()
```

```python
df_dropna.isnull().sum()
```

```python
data_clean = df_dropna
```

## Some basic filtering and grouping

```python
# Question 3: Filter the dataset to include only schools located in Brooklyn
brooklyn_schools = df[df['borough'].str.lower() == 'brooklyn']
print(brooklyn_schools.shape)
```

```python
brooklyn_schools.head()
```

```python
# Question 4a: Total number of school in Brooklyn
total_brooklyn_schools = brooklyn_schools.shape[0]
print(f"Total schools in Brooklyn: {total_brooklyn_schools}")
```

```python
# 4b. How many offer Grade 9 entry?
grade9_entry = brooklyn_schools[brooklyn_schools['grade_span_min'] <= 9]
print(f"Brooklyn schools offering Grade 9 entry or earlier: {grade9_entry.shape[0]}")

```

```python
#Question 5a: Count of school per borough
school_counts = df['borough'].value_counts()
print("Count of schools per borough:\n", school_counts)
```

```python
# Question 5b: Average number of students in Borough
avg_students = df.groupby('borough')['total_students'].mean().round(1)
print("Average number of students per borough:\n", avg_students)
```

```python
#Question 5C: Summary of grade_span_max grouped by borough
grade_summary = df.groupby('borough')['grade_span_max'].describe()
print("Summary of grade_span_max by borough:\n", grade_summary)
```

## Some Simple Visualizations

```python
# Question 6: Number of schools per Borough
sns.countplot(data=df, x='borough', hue='borough')
plt.title("Number of Schools by Borough")
plt.xticks(rotation=45)
plt.show()
```

```python
## Total Students Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['total_students'].dropna(), bins=20, kde=True)
plt.title("Distribution of Total Students")
plt.xlabel("Total Students")
plt.show()
```
