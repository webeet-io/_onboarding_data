```python
# Importing pandas library, loading the data set and checking if everything is ok (head() function)

import pandas as pd
df = pd.read_csv("high-school-directory.csv")
df.head()
```

```python
# Cleaning column names (all lower case, no whitespaces at the start and end of a string,
# replacing space with underscore)

new_column_names=[]

for column in df.columns:
        clean_name=column.lower().strip().replace(' ','_')
        new_column_names.append(clean_name)

new_column_names
```

```python
# Assigning new cleaned column names to the data frame and seeing if it worked

df.columns = new_column_names
df.head()
```

```python
# Filtering the dataset to include only schools located in
# Brooklyn

schools_in_brooklyn = df[df['borough']=='Brooklyn']
schools_in_brooklyn
```

```python
# Using previous variable schools_in_brooklyn counting the number of schools

print('Schools in Brooklyn:', schools_in_brooklyn['dbn'].nunique())
```

```python
# How many offer Grade 9 entry?

# All data set
print('In all New York:', df[(df['grade_span_min'] <= 9) & (df['grade_span_max'] >= 9)].shape[0])

# Only in Brooklyn
print('Only in Brooklyn:', schools_in_brooklyn[(schools_in_brooklyn['grade_span_min'] <= 9) & (schools_in_brooklyn['grade_span_max'] >= 9)].shape[0])
```

```python
# Count of schools per borough?

count_of_schools_per_borough = df.groupby('borough')['dbn'].nunique()
count_of_schools_per_borough
```

```python
# Average number of students per borough?

average_students_per_borough = df.groupby('borough')['total_students'].mean().round(2)
average_students_per_borough
```

```python
# Summary of grade_span_max grouped by borough?

summary_of_grade_span_max_by_borough = df.groupby('borough')['grade_span_max'].describe()
summary_of_grade_span_max_by_borough
```

```python
# Bar chart: Number of schools per borough?
# First importing seaborn, modifing a variable count_of_schools_per_borough to a dataframe and then creating a visualization

import seaborn as sns

schools_per_borough = count_of_schools_per_borough.reset_index()

schools_per_borough_barplot = sns.barplot(x="borough", y="dbn", data=schools_per_borough, palette='colorblind', hue='borough')
schools_per_borough_barplot.set_title('Number of Schools per Borough', fontsize=16)
schools_per_borough_barplot.set_xlabel('')
schools_per_borough_barplot.set_ylabel('Number of Schools', fontsize=14)
```

```python
# Some more insights:
# Which school offers the most programs (Top 5)

top_5_schools_by_programs = df.sort_values(by='number_programs', ascending=False).head(5)
top_5_schools_by_programs[['school_name', 'number_programs']]
```
