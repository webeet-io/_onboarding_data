```python
import pandas as pd

# Load the dataset
df = pd.read_csv("high-school-directory.csv")

# Preview the data
df.head()
```

```python
# Clean column names
df.columns = (
    df.columns
    .str.strip()                      # remove leading/trailing spaces
    .str.lower()                      # make lowercase
    .str.replace(' ', '_')            # replace spaces with underscores
    .str.replace('[^a-z0-9_]', '', regex=True)  # remove special characters
)

# Check the cleaned column names
df.columns
```

```python
print(df.columns)  # show the cleaned column names as a list
df.head()          # show the first few rows of the updated table
```

```python
# Show only schools in Brooklyn
brooklyn_schools = df[df['borough'] == 'Brooklyn']
brooklyn_schools.head()
```

```python
# unique value of borough
df['borough'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of entries per borough
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='borough', order=df['borough'].value_counts().index, palette='viridis')
plt.title('Number of Entries per Borough')
plt.xlabel('Borough')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 most common grade spans (min-max) among schools
df[['grade_span_min', 'grade_span_max']].value_counts().head(10)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 grade spans
top_10_grades = df[['grade_span_min', 'grade_span_max']].value_counts().head(10)

# Convert index (MultiIndex) to strings for plotting
grade_labels = [f"{min_grade}-{max_grade}" for min_grade, max_grade in top_10_grades.index]

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_10_grades.values, y=grade_labels, palette='coolwarm')
plt.title('Top 10 Most Common Grade Spans')
plt.xlabel('Number of Schools')
plt.ylabel('Grade Span (Min-Max)')
plt.show()

```

```python
# Import the libraries I'll need
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Checking how many schools fall under each council district
# (Here I'm just mocking up a small sample to visualize)
districts = range(1, 11)  # Districts 1 through 10
num_schools = [15, 20, 12, 18, 25, 10, 22, 16, 19, 14]

schools_per_district = pd.Series(num_schools, index=districts)

# Plot the number of schools per district
plt.figure(figsize=(10, 5))
sns.barplot(
    x=schools_per_district.index,
    y=schools_per_district.values,
    palette="viridis"
)

plt.title("Number of Schools per Council District", fontsize=14)
plt.xlabel("Council District", fontsize=12)
plt.ylabel("Number of Schools", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
# Find the top 10 neighborhoods with the most schools
df['nta'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get top 10 neighborhoods by number of schools
top_10_nta = df['nta'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_10_nta.values, y=top_10_nta.index, palette='magma')
plt.title('Top 10 Neighborhoods with the Most Schools')
plt.xlabel('Number of Schools')
plt.ylabel('Neighborhood (NTA)')
plt.show()
```
