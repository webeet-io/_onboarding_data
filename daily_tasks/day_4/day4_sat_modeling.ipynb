```python
# Fix inconsistent formatting
df['pct_students_tested'] = df['pct_students_tested'].str.replace('%', '').astype(float)
df
```

```python
# Changing the datatype from "object" to "numeric"
df["SAT Critical Reading Avg. Score"] = pd.to_numeric(df["SAT Critical Reading Avg. Score"], errors="coerce")
df["SAT Writing Avg. Score"] = pd.to_numeric(df["SAT Writing Avg. Score"], errors="coerce")
df["SAT Math Avg. Score"] = pd.to_numeric(df["SAT Math Avg. Score"], errors="coerce")
df["SAT Critical Reading Avg. Score"] = pd.to_numeric(df["SAT Critical Reading Avg. Score"], errors="coerce")
df["pct_students_tested"] = pd.to_numeric(df["pct_students_tested"], errors="coerce")
```

```python
df.describe()
```

```python
df["Num of SAT Test Takers"] = pd.to_numeric(df["Num of SAT Test Takers"], errors="coerce")
df["SAT Critical Readng Avg. Score"] = pd.to_numeric(df["SAT Critical Readng Avg. Score"], errors="coerce")
```

```python
# Replaces those invalid values with NaN (Not a Number)
# SAT Math scores should be between 200 and 800, so anything outside that range is likely bad data.
df.loc[
    (df["SAT Math Avg. Score"] < 200) | (df["SAT Math Avg. Score"] > 800),
    "SAT Math Avg. Score"
] = np.nan
```
