```python
# fixing column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Droping duplicated or typo column
df = df.drop(columns=['sat_critical_readng_avg._score'], errors='ignore')

df.head()
```

```python
# Convert numeric columns from text to numbers
cols_to_convert = [
    'num_of_sat_test_takers',
    'sat_critical_reading_avg._score',
    'sat_math_avg._score',
    'sat_writing_avg._score'
]

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
import pandas as pd

# Ensure all values are strings before using .str methods
df['pct_students_tested'] = (
    df['pct_students_tested']
    .astype(str)                         # convert everything to string first to guarantee easy conversion from 'percent' to 'numeric' as you suggested.
    .str.replace('%', '', regex=False)   # removes % sign
)

# Now convert to numeric (turns missing entries into NaN)
df['pct_students_tested'] = pd.to_numeric(df['pct_students_tested'], errors='coerce')


```

```python
# Fill missing text values
df['contact_extension'] = df['contact_extension'].fillna("Unknown")

# Fill missing numeric values
numeric_cols_mean = [
    'num_of_sat_test_takers',
    'sat_critical_reading_avg._score',
    'sat_math_avg._score',
    'sat_writing_avg._score',
    'pct_students_tested',
    'academic_tier_rating'
]

for col in numeric_cols_mean:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col].fillna(df[col].mean(), inplace=True)

# Confirm there are no missing values now
df.isnull().sum()
```

```python
sat_cols = ['sat_critical_reading_avg._score', 'sat_math_avg._score', 'sat_writing_avg._score']

for col in sat_cols:
    df.loc[(df[col] < 200) | (df[col] > 800), col] = None  # replace invalid scores with NaN
```

```python
for col in sat_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```
