```python
df=pd.read_csv('/Users/msubhanzk/Downloads/2. Data Analysis Course/14. June 2025/1. Internship Documents/4. Day 4/1. Dataset/sat-results.csv')
pd.set_option("Display.max_rows", None)
df
```

```python
## Here I cleaned the columns' names such as 
### Replacing spaces with underscore,
### Convert all columns to lowercase,
### Removing dots
### Removing leading/trailing spaces

df.columns = (
        df.columns
        .str.strip()
        .str.replace(" ", "_")
        .str.replace(r"\.+", "", regex = True)
        .str.lower()
)
print(df.columns.tolist())

df.head()
```

```python
## Replace the string to zero (0) in the specific columns

columns_to_clean = [
    "num_of_sat_test_takers",
    "sat_critical_reading_avg_score",
    "sat_math_avg_score",
    "sat_writing_avg_score",
    "sat_critical_readng_avg_score"
]
for col in columns_to_clean:
    df[col] = pd.to_numeric(df[col], errors= "coerce").fillna(0)
pd.set_option("display.max_rows", None)
display(df)
```

```python
## Replacing NaN values to 0%

df["pct_students_tested"]=df["pct_students_tested"].fillna("0%")
df["academic_tier_rating"]= df["academic_tier_rating"].replace(" ",0).fillna(0)
display(df)
```

```python
## Removed column "contact_extension" because the column will not be required for our dateset.

df = df.drop(columns=["contact_extension"])
display(df)
```

```python
new_df=df.drop_duplicates()
display(new_df)
```

```python
## Upload the clean dataframe to database (PostgreSQL)

new_df.to_sql("sat_results", engine, if_exists="replace", index=False)
```

```python
## checking tables in db
query="""
SELECT table_schema, table_name
FROM information_schema.tables
WHERE table_type = 'BASE TABLE'
  AND table_schema NOT IN ('pg_catalog', 'information_schema')
ORDER BY table_schema, table_name;"""
df = pd.read_sql(query, engine)
df
```

```python
## checking columns
query="""
SELECT 
    table_schema, 
    table_name, 
    column_name, 
    data_type
FROM information_schema.columns
WHERE table_name = 'sat_results'
ORDER BY ordinal_position;"""
df = pd.read_sql(query, engine)
df
```
